```python
import matplotlib as plt
import numpy as np
import seaborn as sns
import pandas as pd

df=pd.read_csv('C:/Users/DHYEY/OneDrive/Desktop/project.csv')
df
```

```python
#Data Visualization using Seaborn
#Line Plot
sns.lineplot( df[ 'recency'], df[ 'complaints' ])
```

```python
#Historgram
sns.distplot(df['complaints' ].head(10))
```

```python
sns.distplot(df[ 'complaints' ].head(30),kde=False,bins=10)
```

```python
#Box Plot
sns.boxplot( df['recency'].head(20))
```

```python
#Bar Plot
sns.barplot(df[ 'complaints'].head(5), df['recency'].head(5))
```

```python
#violin Plot
sns.violinplot(df['recency'].head(10))
```

```python
#Swarm Plot
sns.swarmplot(df['recency'].head(30), df['complaints'].head(30))
```

```python
#Point Plot
sns.pointplot(df['recency'].head(10), df['complaints'].head(10))
```
